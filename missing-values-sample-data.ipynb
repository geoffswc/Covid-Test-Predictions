```python
import pandas as pd
```

```python
headers = ['R', 'temperature', 'cough_severity', 'difficult_breathing', 'exposure_risk', 'days_since_onset']
```

```python
rows = [
    (1, 101.2, 3, None, 1, 15), 
    (1, None, 1, 0, 1, 7), 
    (0, 98.5, 1, 0, 1, 0), 
    (1, 99.0, 1, 0, 1, None), 
    (0, 98.5, None, 0, 0, 21), 
    (1, 99.0, 1, 0, 1, 10), 
    (0, None, 1, 0, 1, 5), 
    (0, 98.2, 2, 1, None, 4)]
```

```python
df = pd.DataFrame(rows, columns=headers)
```

```python
df
```

```python
#https://pandas.pydata.org/pandas-docs/stable/user_guide/missing_data.html
```

```python
df.dropna()
```

```python
df.fillna(0)
```

```python
df.fillna(df.mean()).round(2)
```

```python
# interpolate with back fill
df.interpolate().bfill()
```

```python
#https://scikit-learn.org/stable/modules/impute.html#impute
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
nan = np.nan

imp = SimpleImputer(missing_values=np.nan, strategy='mean')

imp.fit(df)

pd.DataFrame(imp.transform(df), columns=headers).round(2)
```

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=2, weights="uniform")
pd.DataFrame(imputer.fit_transform(df), columns=headers)
```
