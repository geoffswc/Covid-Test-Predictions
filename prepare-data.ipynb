```python
import pandas as pd
from pandasql import sqldf
import glob
pysqldf = lambda q: sqldf(q, globals())
```

```python
all_files = glob.glob('/Users/geoffswc/covidclinicaldata/data/*.csv')
df = pd.concat((pd.read_csv(f) for f in all_files), sort=False)
```
